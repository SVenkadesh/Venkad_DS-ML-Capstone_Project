```python
# Import neccessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import Dataset
dataset = pd.read_csv('Indian_Kids_Screen_Time.csv')
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset['Health_Impacts'] = dataset['Health_Impacts'].fillna('Not Impact')
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset['Exceeded_Recommended_Limit'] = dataset['Exceeded_Recommended_Limit'].astype('int64')

map_Gender = {'Male':1,'Female':0}
map_device = {'Smartphone':0,'Tablet':1,'Laptop':2,'TV':3}
map_urban_rural = {'Urban':0,'Rural':1}
map_health_impact = {"Poor Sleep, Eye Strain":0,"Poor Sleep":1,"Not Impact":2,"Poor Sleep, Anxiety":3,"Poor Sleep, Obesity Risk":4,"Eye Strain":5,"Obesity Risk":6,
"Anxiety":7,"Poor Sleep, Anxiety, Obesity Risk":8,"Eye Strain, Obesity Risk":9,"Eye Strain, Anxiety, Obesity Risk":10,"Anxiety, Obesity Risk":11,
"Poor Sleep, Eye Strain, Obesity Risk":12,"Poor Sleep, Eye Strain, Anxiety":13,"Poor Sleep, Eye Strain, Anxiety, Obesity Risk":14,"Eye Strain, Anxiety":15 }

dataset['Gender'] = dataset['Gender'].replace(map_Gender)
dataset['Primary_Device'] = dataset['Primary_Device'].replace(map_device)
dataset['Urban_or_Rural'] = dataset['Urban_or_Rural'].replace(map_urban_rural)
dataset['Health_Impacts'] = dataset['Health_Impacts'].replace(map_health_impact)
```

```python
dataset.info()
```

```python
dataset
```

```python
dataset.describe()
```

```python
dataset.columns
```

```python
input = dataset[['Age', 'Gender', 'Avg_Daily_Screen_Time_hr', 'Primary_Device',
       'Exceeded_Recommended_Limit', 'Educational_to_Recreational_Ratio', 'Urban_or_Rural']]

output = dataset[['Health_Impacts']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(input,output,test_size=1/3,random_state=0)
```

```python
# Standard Scaler

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# SVMC

from sklearn.svm import SVC
classifier=SVC(kernel='rbf',random_state=0)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
classifier = confusion_matrix(y_test,y_pred,normalize='true')
classifier
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test,y_pred,output_dict=True)
```
